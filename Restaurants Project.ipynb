### Find out the ratio between restaurants that allow table booking vs. those that do not allow table booking

```python
Booking_table=df.groupby('Has_Table_booking').Restaurant_ID.count()
Booking_table
```

```python
print('Ratio between restaurants that allow table booking vs. those that do not allow table booking:', 
      round((Booking_table.Yes/Booking_table.No),3))
```

```python
plt.title('Booking Availability')
plt.pie(Booking_table,labels=('Table Booking Not Allowed','Table Booking Allowed'),explode=(0.1,0),shadow=True,
        autopct='%1.2f%%')
plt.show()
```

###### The chart above shows that only 12.13% restaurants allow table booking and 87.87% don't allow table booking.

```python
Cuisine3_count=pd.DataFrame(cuisines_split["Cuisine 3"].value_counts()).reset_index()
Cuisine3_count=Cuisine3_count.rename(columns={"index": "Cuisines", "Cuisine 3": "Counts"})
Cuisine3_count.head(6)
```

```python
Cuisine4_count=pd.DataFrame(cuisines_split["Cuisine 4"].value_counts()).reset_index()
Cuisine4_count=Cuisine4_count.rename(columns={"index": "Cuisines", "Cuisine 4": "Counts"})
Cuisine4_count.head(6)
```

```python
Cuisine5_count=pd.DataFrame(cuisines_split["Cuisine 5"].value_counts()).reset_index()
Cuisine5_count=Cuisine5_count.rename(columns={"index": "Cuisines", "Cuisine 5": "Counts"})
Cuisine5_count.head(6)
```

```python
Rating_corr=df[['Restaurant_ID','Restaurant_Name','Country','City','Aggregate_rating','Average_Cost_for_two','Votes',
                'Price_range','Has_Table_booking','Has_Online_delivery']]
```

```python
dummy = ['Has_Table_booking','Has_Online_delivery'] # 0 indicates 'NO' and 1 indicates 'YES'
Rating_corr = pd.get_dummies(Rating_corr,columns=dummy,drop_first=True)
Rating_corr = Rating_corr.merge(CuisineCount_by_res,left_on='Restaurant_Name',right_on='Restaurant_Name',how='left')
Rating_corr.head()
```

##### Correlation between "Rating" and "Average cost for two"

```python
sns.regplot(x='Average_Cost_for_two',y='Aggregate_rating',data=Rating_corr)
Rating_corr[["Average_Cost_for_two", "Aggregate_rating"]].corr()
```

###### The charts above shows weak positive correlation between Average cost for two and Rating of the restaurants.

### Correlation between "Rating" and "Has Table Booking" - 1 indicates 'YES' and 0 indicates 'NO'

```python
sns.regplot(x='Has_Table_booking_Yes',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Has_Table_booking_Yes','Aggregate_rating']].corr()
```

###### This shows a good positive correlation between table booking option and Rating and could be one of the basis for good rating

### Correlation between "Rating" and "Has Online delivery" - 1 indicates 'YES' and 0 indicates 'NO'

```python
sns.regplot(x='Has_Online_delivery_Yes',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Has_Online_delivery_Yes','Aggregate_rating']].corr()
```

###### There is a good positive correlation between online delivry option and Rating. This could be another big factor for giving a good rating to a restaurant

### Correlation between "Rating" and "Price Range"

```python
sns.regplot(x='Price_range',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Price_range','Aggregate_rating']].corr()
```

###### Price range has very positive correlation with rating, as with increasing price range rating also gets better.

### Correlation between "Rating" and "Votes"

```python
sns.regplot(x='Votes',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Votes','Aggregate_rating']].corr()
```

###### Votes also have a very positive correlation with rating, as with increasing price range rating also gets better.

### Correlation between "Rating" and "Cuisines Count"

```python
sns.regplot(x='Cuisines',y='Aggregate_rating',data=Rating_corr)
Rating_corr[['Cuisines','Aggregate_rating']].corr()
```

###### The Number of Cuisines provided by a restaurant is worst correlated to Rating. There is very minimal relation. Hence this cannot be the deciding factor for Rating.
